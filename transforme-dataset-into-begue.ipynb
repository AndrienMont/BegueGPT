# On applique pour chaque élement du jeu de donnée un fonction qui transforme la réponse en bagayement

```python
import random
import re
import time

def begue(mot):
    repetitions = random.randint(1, 3)  # Nombre aléatoire de répétitions (1 à 3)

    if len(mot) > 1:
        if mot[1] in "aeiouy":
            return '-'.join([mot[:2]] * repetitions) + '-' + mot[2:]
        else:
            return '-'.join([mot[0]] * repetitions) + '-' + mot
    else:
        return mot

def troncate(text, max_length):
    return text if len(text) <= max_length else text[:max_length]

def process_text(input_text):
    words = re.findall(r'\b\w+\b', input_text)
    processed_words = []
    for word in words:
        if word.isalpha() and random.random() < 0.5:
            processed_words.append(begue(word))
        else:
            processed_words.append(word)
    return ' '.join(processed_words)

def transform_text(input_text):
    text = input_text["output"]
    message = {
        "output": process_text(text)
    }

    return message
```

```python
# ds["train"]
begue_ds = ds.map(transform_text)
```
